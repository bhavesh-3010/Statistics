# Statistics

1. What is hypothesis testing in statistics?

     Hypothesis testing in statistics is a method used to make inferences about a population based on sample data by testing an assumption, called the null hypothesis (H‚ÇÄ), which typically states that there is no effect or difference. An alternative hypothesis (H‚ÇÅ) represents what you aim to support. Using a test statistic calculated from the sample, you determine the likelihood (p-value) of observing the data if H‚ÇÄ is true. If the p-value is less than a predetermined significance level (commonly 0.05), you reject H‚ÇÄ, suggesting there is enough evidence to support H‚ÇÅ. Otherwise, you fail to reject H‚ÇÄ, meaning the data doesn‚Äôt provide strong enough evidence against it.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis

     - The null hypothesis (H‚ÇÄ) is a statement that assumes there is no effect, no difference, or no relationship between variables‚Äîit represents the default or status quo assumption in a hypothesis test. In contrast, the alternative hypothesis (H‚ÇÅ or Ha) is what you aim to prove; it suggests that there is an effect, a difference, or a relationship. The null hypothesis is tested directly, and based on the evidence from the sample data, it is either rejected in favor of the alternative hypothesis or not rejected if the data does not provide strong enough support.

3. What is the significance level in hypothesis testing, and why is it important?

      - The significance level, denoted by alpha (Œ±), is the threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the probability of making a **Type I error**, which is rejecting the null hypothesis when it is actually true. Common values for Œ± are 0.05, 0.01, or 0.10. The significance level is important because it sets the standard for how strong the evidence must be to conclude that the observed effect is real and not due to random chance. Choosing an appropriate Œ± helps balance the risks of making incorrect conclusions in statistical decision-making.

4. What does a P-value represent in hypothesis testing?

     - In hypothesis testing, a **p-value** represents the probability of obtaining the observed results, or more extreme ones, assuming that the **null hypothesis is true**. It measures the strength of evidence against the null hypothesis: the smaller the p-value, the stronger the evidence to reject H‚ÇÄ. If the p-value is less than or equal to the chosen significance level (e.g., 0.05), you reject the null hypothesis, suggesting that the observed effect is statistically significant. A large p-value means the data is consistent with H‚ÇÄ, and there is not enough evidence to support the alternative hypothesis.

5. How do you interpret the P-value in hypothesis testing

      - To interpret the p-value in hypothesis testing, compare it to the chosen significance level (Œ±), usually 0.05. If the **p-value is less than or equal to Œ±**, it means the observed results are unlikely to have occurred by random chance under the null hypothesis, so you **reject the null hypothesis** and conclude there is statistically significant evidence for the alternative hypothesis. If the **p-value is greater than Œ±**, you **fail to reject the null hypothesis**, meaning there isn‚Äôt enough evidence to support the alternative, and the results could be due to chance. The smaller the p-value, the stronger the evidence against the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?

  - A Type 1 error occurs when the null hypothesis (H‚ÇÄ) is true, but you wrongly reject it‚Äîthis is also known as a "false positive." A Type 2 error happens when the null hypothesis is false, but you fail to reject it‚Äîcalled a "false negative." Type 1 errors are controlled by the significance level (Œ±), while Type 2 errors are related to the test‚Äôs power. Balancing both is essential to minimize wrong conclusions in hypothesis testing.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
¬†¬†- A one-tailed test checks for an effect in only one direction (either greater than or less than), while a two-tailed test checks for effects in both directions (not equal to). One-tailed tests are more powerful in detecting a specific direction of change, but two-tailed tests are used when the direction of the effect is not known or when deviations in both directions are important.

8. What is the Z-test, and when is it used in hypothesis testing?
¬†¬†- A Z-test is used to determine whether there is a significant difference between sample and population means when the population standard deviation is known and the sample size is large. It assumes the data is normally distributed and is often used for large-scale data analysis or when population parameters are available.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
¬†¬†- The T-distribution is used when the sample size is small (typically less than 30) and/or the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails to account for greater variability. It provides more accurate results in small-sample hypothesis testing.



11. What is the difference between a Z-test and a T-test?
¬†¬†- A Z-test is used when the population standard deviation is known and the sample size is large, while a T-test is used when the standard deviation is unknown and/or the sample size is small. Z-tests use the standard normal distribution; T-tests use the t-distribution, which adjusts for extra uncertainty in small samples.

12. What is the T-test, and how is it used in hypothesis testing?
¬†¬†- A T-test is used to compare means and determine if they are significantly different. It is used when the population standard deviation is unknown. Types include one-sample, two-sample (independent), and paired-sample t-tests, and they are used based on whether you're comparing one group, two independent groups, or two related groups.

13. What is the relationship between Z-test and T-test in hypothesis testing?
¬†¬†- Both Z-tests and T-tests assess whether differences between means are statistically significant. The key difference lies in sample size and knowledge of population standard deviation. As sample size increases, the t-distribution approaches the normal distribution, and results from both tests become nearly identical.

14. What is a confidence interval, and how is it used to interpret statistical results?
¬†¬†- A confidence interval (CI) is a range of values derived from sample data that is likely to contain the true population parameter with a specified level of confidence (e.g., 95%). It provides an estimate of uncertainty around a sample statistic. A narrower interval indicates more precision, while a wider one suggests more uncertainty.

15. What is the margin of error, and how does it affect the confidence interval?
¬†¬†- The margin of error is the amount added to and subtracted from the sample statistic to create a confidence interval. It reflects the maximum expected difference between the true population parameter and the sample estimate. A larger margin results in a wider confidence interval, indicating more uncertainty in the estimate.

16. How is Bayes' Theorem used in statistics, and what is its significance?
¬†¬†- Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It combines prior probability with the likelihood of observed data to produce a revised (posterior) probability. It is significant in decision-making processes under uncertainty and widely used in fields like medicine, machine learning, and economics.



17. What is the Chi-square distribution, and when is it used?
¬†¬†- The Chi-square distribution is a skewed distribution used in tests involving categorical data, such as goodness-of-fit tests and tests of independence. It is based on the sum of squared standard normal variables and is only defined for non-negative values. Its shape depends on degrees of freedom.



18. What is the Chi-square goodness of fit test, and how is it applied?
¬†¬†- The Chi-square goodness of fit test assesses whether observed categorical data matches expected distributions. It is applied by comparing observed frequencies to expected ones under a given hypothesis. If the calculated chi-square statistic exceeds the critical value, the null hypothesis (that the data fits the distribution) is rejected.

19. What is the F-distribution, and when is it used in hypothesis testing?
¬†¬†- The F-distribution is used in tests that compare two or more sample variances, such as in ANOVA and regression analysis. It is positively skewed and defined only for positive values. The shape depends on two sets of degrees of freedom. A large F-value indicates greater differences between group variances or model fits.

20. What is an ANOVA test, and what are its assumptions?
¬†¬†- ANOVA (Analysis of Variance) is used to compare the means of three or more groups to determine if at least one group mean differs significantly. Its assumptions include: independence of observations, normality of residuals, and equal variances across groups. If assumptions are violated, the results may not be valid.

21. What are the different types of ANOVA tests?
¬†¬†- The main types of ANOVA tests are:
¬†¬†¬†¬†- One-way ANOVA: compares means across one factor.
¬†¬†¬†¬†- Two-way ANOVA: examines effects of two factors and their interaction.
¬†¬†¬†¬†- Repeated Measures ANOVA: compares means across multiple time points or conditions within the same subjects.
¬†¬†- Each type is suited to different experimental designs.

22. What is the F-test, and how does it relate to hypothesis testing?
¬†¬†- The F-test is a statistical test that compares the ratio of two variances to determine if they are significantly different. It is used in ANOVA and regression to test whether groups have equal variances or whether a model explains a significant portion of variance. A high F-value relative to the critical value leads to rejection of the null hypothesis.

```python
#1Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy import stats

# Sample data
sample = [52, 55, 49, 51, 50, 53, 48, 54, 56, 50]  # Sample data
population_mean = 50  # Known population mean
population_std = 3    # Known population standard deviation (œÉ)
alpha = 0.05          # Significance level

# Calculate sample statistics
sample_mean = np.mean(sample)
n = len(sample)
standard_error = population_std / np.sqrt(n)

# Calculate Z-score
z_score = (sample_mean - population_mean) / standard_error

# Calculate P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Interpretation
print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between the sample and population mean.")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample and population details
sample = [102, 98, 101, 100, 99, 97, 103, 104, 96, 100]
population_mean = 100
population_std = 3
alpha = 0.05

# Z-test calculations
sample_mean = np.mean(sample)
n = len(sample)
standard_error = population_std / np.sqrt(n)
z_score = (sample_mean - population_mean) / standard_error
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print Z-test results
print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
else:
    print("Result: Fail to reject the null hypothesis.")

# Visualization of the decision region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Critical Z values for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5)

# Plot the Z-score
plt.axvline(z_score, color='black', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='green', linestyle=':', linewidth=2, label=f'Z-critical = ¬±{z_critical:.2f}')
plt.axvline(z_critical, color='green', linestyle=':', linewidth=2)

# Labels and legend
plt.title("Two-tailed Z-test Visualization")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=102, sigma=3, n=30, alpha=0.05):
    """
    Visualize Type I and Type II errors for a two-tailed Z-test.
    mu0: Mean under the null hypothesis
    mu1: Mean under the alternative hypothesis
    sigma: Standard deviation of population
    n: Sample size
    alpha: Significance level
    """
    se = sigma / np.sqrt(n)
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    # Null and alternative distributions
    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    # Critical values for rejection region (two-tailed)
    z_crit = norm.ppf(1 - alpha/2)
    lower_crit = mu0 - z_crit * se
    upper_crit = mu0 + z_crit * se

    # Plot the distributions
    plt.figure(figsize=(12, 6))
    plt.plot(x, y_null, label='Null Hypothesis (H‚ÇÄ)', color='blue')
    plt.plot(x, y_alt, label='Alternative Hypothesis (H‚ÇÅ)', color='green')

    # Shade Type I error (alpha)
    plt.fill_between(x, y_null, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Shade Type II error (beta)
    plt.fill_between(x, y_alt, where=(x >= lower_crit) & (x <= upper_crit), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    # Vertical lines
    plt.axvline(lower_crit, color='black', linestyle='--', label=f'Lower Critical Value: {lower_crit:.2f}')
    plt.axvline(upper_crit, color='black', linestyle='--', label=f'Upper Critical Value: {upper_crit:.2f}')

    # Labels and legend
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function to visualize
visualize_type1_type2_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean using the t-distribution.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default 0.95 for 95%)

    Returns:
        Tuple (lower_bound, upper_bound)
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    se = sample_std / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * se
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

# Example usage
sample = [10, 12, 13, 11, 14, 12, 13, 15, 10, 11]
ci = confidence_interval(sample, confidence=0.95)
print(f"95% Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default 0.95 for 95%)

    Returns:
        Margin of error (float)
    """
    data = np.array(data)
    n = len(data)
    se = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    moe = t_crit * se
    return round(moe, 3)

# Example usage
sample_data = [52, 55, 53, 50, 54, 56, 52, 53, 55, 51]
confidence = 0.95
moe = margin_of_error(sample_data, confidence)

print("Sample Mean:", np.mean(sample_data))
print(f"{int(confidence*100)}% Margin of Error:", moe)
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, sensitivity, specificity):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): P(Disease) ‚Äî prior probability of having the disease
        sensitivity (float): P(Positive | Disease) ‚Äî true positive rate
        specificity (float): P(Negative | No Disease) ‚Äî true negative rate

    Returns:
        posterior (float): P(Disease | Positive) ‚Äî probability of disease given a positive test
    """
    # Complement values
    p_disease = prior
    p_no_disease = 1 - prior
    p_pos_given_disease = sensitivity
    p_pos_given_no_disease = 1 - specificity

    # Total probability of a positive test
    p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Bayes' Theorem
    posterior = (p_pos_given_disease * p_disease) / p_positive
    return round(posterior, 4)

# Example scenario
prior = 0.01           # 1% of population has the disease
sensitivity = 0.95     # 95% true positive rate
specificity = 0.98     # 98% true negative rate

posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Probability of having the disease given a positive test result: {posterior * 100:.2f}%")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Simulated survey data: Gender vs Preferred Product
data = {'Product A': [20, 15],
        'Product B': [30, 35],
        'Product C': [25, 20]}
index = ['Male', 'Female']

# Create a contingency table
contingency_table = pd.DataFrame(data, index=index)

# Perform Chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Contingency Table:")
print(contingency_table, '\n')
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}\n")

# Interpret the result
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis ‚Äî the variables are dependent (not independent).")
else:
    print("Result: Fail to reject the null hypothesis ‚Äî no evidence of dependency between the variables.")
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., from a dice roll experiment)
observed = np.array([18, 22, 20, 17, 19, 24])  # 6 categories (like die faces)

# Expected frequencies (e.g., fair die ‚Üí uniform distribution)
expected = np.array([20] * 6)  # All outcomes equally likely

# Perform the chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the null hypothesis ‚Äî the data does not follow the expected distribution.")
else:
    print("\nResult: Fail to reject the null hypothesis ‚Äî no significant difference from the expected distribution.")
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Perform the F-test (larger variance / smaller variance)
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

# Output results
print("Sample 1 Variance:", round(var1, 2))
print("Sample 2 Variance:", round(var2, 2))
print("F-statistic:", round(f_stat, 4))
print("Degrees of Freedom:", df1, "and", df2)
print("P-value:", round(p_value, 4))

# Interpret result
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the null hypothesis ‚Äî the variances are significantly different.")
else:
    print("\nResult: Fail to reject the null hypothesis ‚Äî no significant difference in variances.")
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import shapiro, levene
from statsmodels.stats.stattools import durbin_watson

def check_anova_assumptions(data, group_col, value_col):
    """
    Check ANOVA assumptions: Normality, Homogeneity of variance, Independence.

    Parameters:
    - data: Pandas DataFrame with grouped data
    - group_col: Name of the column indicating group
    - value_col: Name of the column with numerical values

    Returns:
    - Prints test statistics and p-values for assumptions
    """
    # Split into groups
    groups = [group[value_col].values for name, group in data.groupby(group_col)]

    # 1. Normality (Shapiro-Wilk on residuals)
    model = sm.formula.ols(f"{value_col} ~ C({group_col})", data=data).fit()
    residuals = model.resid
    stat, p_normal = shapiro(residuals)
    print("1. Shapiro-Wilk Normality Test:")
    print(f"   Statistic = {stat:.4f}, p-value = {p_normal:.4f}")
    print("   " + ("Residuals are normal ‚úÖ" if p_normal > 0.05 else "Residuals are NOT normal ‚ùå"))

    # 2. Homogeneity of variance (Levene‚Äôs test)
    stat, p_levene = levene(*groups)
    print("\n2. Levene's Test for Equal Variances:")
    print(f"   Statistic = {stat:.4f}, p-value = {p_levene:.4f}")
    print("   " + ("Variances are equal ‚úÖ" if p_levene > 0.05 else "Variances are NOT equal ‚ùå"))

    # 3. Independence (Durbin-Watson on residuals)
    dw_stat = durbin_watson(residuals)
    print("\n3. Durbin-Watson Test for Independence:")
    print(f"   Statistic = {dw_stat:.4f}")
    if 1.5 < dw_stat < 2.5:
        print("   Residuals are independent ‚úÖ")
    else:
        print("   Residuals may not be independent ‚ùå")

# Example usage with simulated data
if __name__ == "__main__":
    np.random.seed(42)
    df = pd.DataFrame({
        'Group': ['A']*30 + ['B']*30 + ['C']*30,
        'Score': np.concatenate([
            np.random.normal(70, 5, 30),
            np.random.normal(72, 5, 30),
            np.random.normal(75, 5, 30)
        ])
    })

    check_anova_assumptions(df, 'Group', 'Score')
```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulate data: Study effect of 'Treatment' and 'Gender' on 'Score'
np.random.seed(42)
n = 20

treatments = ['A', 'B']
genders = ['Male', 'Female']
data = []

for treatment in treatments:
    for gender in genders:
        # Simulate some interaction effect
        base = 70 if treatment == 'A' else 75
        gender_effect = 3 if gender == 'Female' else 0
        scores = np.random.normal(loc=base + gender_effect, scale=4, size=n)
        for score in scores:
            data.append([treatment, gender, score])

# Create DataFrame
df = pd.DataFrame(data, columns=['Treatment', 'Gender', 'Score'])

# Two-way ANOVA with interaction
model = ols('Score ~ C(Treatment) + C(Gender) + C(Treatment):C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print results
print("Two-Way ANOVA Results:\n")
print(anova_table)

# Interpretation
print("\nInterpretation:")
for factor in anova_table.index:
    p = anova_table.loc[factor, "PR(>F)"]
    if p < 0.05:
        print(f" - {factor}: Significant effect (p = {p:.4f}) ‚úÖ")
    else:
        print(f" - {factor}: No significant effect (p = {p:.4f}) ‚ùå")

# Visualization
plt.figure(figsize=(8, 6))
sns.pointplot(x='Treatment', y='Score', hue='Gender', data=df, dodge=True, capsize=0.1, errwidth=1.5)
plt.title("Interaction Plot: Treatment √ó Gender")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate two normal distributions
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=100)  # mean=50
sample2 = np.random.normal(loc=52, scale=5, size=100)  # mean=52

# Perform independent t-test
t_stat, p_value = ttest_ind(sample1, sample2)

# Print results
print("Independent T-Test Results:")
print(f"T-statistic = {t_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis ‚Äì the means are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis ‚Äì no significant difference in means.")

# Visualize the distributions
plt.figure(figsize=(8, 5))
sns.kdeplot(sample1, label="Sample 1 (mean=50)", fill=True)
sns.kdeplot(sample2, label="Sample 2 (mean=52)", fill=True)
plt.axvline(np.mean(sample1), color='blue', linestyle='--')
plt.axvline(np.mean(sample2), color='orange', linestyle='--')
plt.title("Comparison of Two Simulated Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
from scipy.stats import chi2

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=100, scale=10, size=30)  # true std dev = 10
sample_size = len(sample)
sample_variance = np.var(sample, ddof=1)

# Known population variance (hypothesized)
pop_variance = 100  # std dev = 10

# Chi-square test statistic
chi2_stat = (sample_size - 1) * sample_variance / pop_variance

# Degrees of freedom
df = sample_size - 1

# Significance level
alpha = 0.05

# Critical values (two-tailed test)
chi2_critical_low = chi2.ppf(alpha / 2, df)
chi2_critical_high = chi2.ppf(1 - alpha / 2, df)

# Print test results
print("Chi-Square Test for Population Variance")
print(f"Sample Variance = {sample_variance:.2f}")
print(f"Chi-Square Statistic = {chi2_stat:.2f}")
print(f"Chi-Square Critical Values (Œ± = {alpha}): [{chi2_critical_low:.2f}, {chi2_critical_high:.2f}]")

# Decision
if chi2_stat < chi2_critical_low or chi2_stat > chi2_critical_high:
    print("Result: Reject the null hypothesis ‚Äì sample variance significantly differs from population variance.")
else:
    print("Result: Fail to reject the null hypothesis ‚Äì no significant difference in variance.")

```

```python
# 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Simulate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=0, scale=5, size=50)
data2 = np.random.normal(loc=0, scale=8, size=50)

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# F-test statistic (larger variance / smaller variance)
f_stat = var2 / var1 if var2 > var1 else var1 / var2

# Degrees of freedom
df1 = len(data1) - 1
df2 = len(data2) - 1

# Significance level
alpha = 0.05

# Critical values
f_crit_low = f.ppf(alpha / 2, df1, df2)
f_crit_high = f.ppf(1 - alpha / 2, df1, df2)

# Print test results
print("F-Test for Comparing Variances")
print(f"Variance 1 = {var1:.4f}")
print(f"Variance 2 = {var2:.4f}")
print(f"F-statistic = {f_stat:.4f}")
print(f"Critical values (Œ± = {alpha}): [{f_crit_low:.4f}, {f_crit_high:.4f}]")

# Interpretation
if f_stat < f_crit_low or f_stat > f_crit_high:
    print("Result: Reject the null hypothesis ‚Äì variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis ‚Äì no significant difference in variances.")

# Visualization
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='F-distribution', color='blue')
plt.axvline(f_stat, color='red', linestyle='--', label='F-statistic')
plt.axvline(f_crit_low, color='green', linestyle='--', label='Lower critical value')
plt.axvline(f_crit_high, color='green', linestyle='--', label='Upper critical value')
plt.fill_between(x, y, where=(x < f_crit_low) | (x > f_crit_high), color='orange', alpha=0.3, label='Rejection regions')
plt.title('F-Test for Comparing Variances')
plt.xlabel('F-value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulated observed data (e.g., categories A‚ÄìD)
observed = np.array([50, 30, 15, 5])

# Expected frequencies under uniform distribution
expected = np.array([25, 25, 25, 25])  # Null: each category has equal chance

# Perform Chi-square goodness of fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Chi-Square Goodness of Fit Test")
print(f"Observed: {observed}")
print(f"Expected: {expected}")
print(f"Chi-square Statistic = {chi2_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis ‚Äì the observed distribution does not fit the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis ‚Äì the observed distribution fits the expected distribution.")

# Visualization
categories = ['A', 'B', 'C', 'D']
x = np.arange(len(categories))

plt.figure(figsize=(8, 5))
bar_width = 0.35
plt.bar(x - bar_width/2, observed, bar_width, label='Observed', color='skyblue')
plt.bar(x + bar_width/2, expected, bar_width, label='Expected', color='orange')
plt.xticks(x, categories)
plt.title('Chi-Square Goodness of Fit Test')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
